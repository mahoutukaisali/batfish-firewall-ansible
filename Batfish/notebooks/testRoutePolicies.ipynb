```python
'''
送信元から宛先までの疎通確認及び経路情報を表示。headersには宛先のホスト名が入る
指定されたフローのパスを、逆フローのパスとともにトレースする。
一方向からの接続性を評価しているので、逆方向からの接続性については見ていない。(二つの方向を分離して、接続の問題をデバックする)

'''
headers = HeaderConstraints(dstIps='as3border2')
tracert = bfq.traceroute(startLocation="as1border1[Loopback0]", headers=headers, maxTraces=1, ignoreFilters=False).answer().frame()
show(tracert)
```

```python
'''
送信元から宛先までの疎通確認及び経路情報を表示。headersには宛先のホスト名が入る。ignoreFiltersをTrueにするとFilterが無視された結果が
返り、Falseにすると投入されているFilter付きで検査される
maxTracesは表示するルートの数を指定する
トレースルートが成功した場合、逆方向からも実行される
'''
headers = HeaderConstraints(dstIps='host1')
tracert = bfq.bidirectionalTraceroute(startLocation="as1border1[Loopback0]", headers=headers, ignoreFilters=True).answer().frame()
show(tracert)
```

```python
# 伝搬されるプレフィックスをトレースする（ルーティング中にNW内のデバイスがプレフィックスを処理する方法を示す）
#最大受信ルート数の制限をONにすると、デフォルトでは75%をこえる経路を受信するとwarning, そして最大受信経路数を超えるとBGP peer disable
#Node=Action発生源。Peer=アクションが適用されるノードのネイバー。
#as1border1のprefix-list(一部抜粋)：ip prefix-list inbound_route_filter seq 5 deny 1.0.0.0/8 le 32
#as1border1がas2border1に送信する1.1.1.1/32はprefix-listの着信トラフィックフィルタによりフィルタリングされる
#Node = the node where action takes place
#Peer = The node's neighbor to which the action applies
'''
Actionで表示されるノードに対して、Prefix列に出ているプレフィックスがNodeから伝搬されるorされない、又は受け取っているのかを表示
ex.)BGPは伝搬するprefixをnetworkコマンドで指定しないとダメだが1.1.1.1/32は指定されていないのでeBGPにもiBGPにも伝搬できていない
'''
question = bfq.prefixTracer(nodes='as1border1').answer().frame()
show(question)
```
