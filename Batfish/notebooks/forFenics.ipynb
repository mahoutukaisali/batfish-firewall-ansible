```python
import json
import csv

ip_owners_ans = bfq.lpmRoutes(ip="1.1.1.1", nodes="as1border1").answer()

frames = ip_owners_ans.frame()

json_frame = frames.to_json()

employee_parsed = json.loads(json_frame)


employ_data = open('longest_match_from_RIB_tutorial.csv', mode='w')
csvwriter = csv.writer(employ_data)

for key, value in employee_parsed.items():
    csvwriter.writerow(key)
    rows = csvwriter.writerow(value)
    print(rows)


#a = csvwriter.writerows((frames.to_csv()))

frames.to_csv('longest_match_from_RIB_tutorial.csv')

employ_data.close()
```
